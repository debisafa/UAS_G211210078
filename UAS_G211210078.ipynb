```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib .dates as dts
import datetime
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/debisafa/UAS_G211210078/main/Dataset_UAS/curah_hujan.csv")
data
```

```python
# Mengatur kolom y
y1 = data['1981-1990']
y2 = data['1991-2000']
y3 = data['2001-2010']

# Membuat array x yang berisi urutan angka dari 0 hingga panjang data pada kolom "Bulan" dalam data
x_ticks = np.arange(len(data['Bulan']))

# Mengubah nama-nama bulan menjadi urutan nomor dari nomor 1
bulan = [str(i + 1) for i in range(len(data['Bulan']))]

# Memberikan ukuran pada diagram batang
width = 0.27
spacing = 0

# Memberikan ukuran figure dengan lebar 10 inch dan tinggi 5 inch
plt.figure(figsize=(10, 5))

# Membuat diagram batang dengan menggunakan data dari kolom '1981-1990', '1991-2000', dan '2001-2010'
bar1 = plt.bar(x_ticks - (width + spacing), y1, color = "b", width = width, label="1981-1990")
bar2 = plt.bar(x_ticks, y2, color = "r", width = width, label="1991-2000")
bar3 = plt.bar(x_ticks + (width + spacing), y3, color = "g", width = width, label="2001-2010")

# Memberi keterangan pada grafik
plt.xlabel('Bulan', fontweight='bold', fontsize=13)
plt.ylabel('Curah Hujan (mm)', fontweight='bold', fontsize=13)
plt.legend(loc="upper center", ncol=3)
plt.xticks(x_ticks, Bulan)
plt.grid(zorder=0, color="gray", linestyle="dashed", linewidth=0.3)

# Menampilkan grafik
plt.show()
```

```python
dt = pd.read_csv("https://raw.githubusercontent.com/debisafa/UAS_G211210078/main/Dataset_UAS/suhu_permukaan.csv")
dt
```

```python
# Membuat array tanggal dan waktu dari tanggal '2013-07-24' dengan interval setiap 3 jam
date = pd.date_range(start='2013-07-24', periods=9, freq='3H')

# Mengambil kolom untuk dijadikan chart
asimilasi = dt['Asimilasi']
observasi = dt['Observasi']
non_asimilasi = dt['Non Asimilasi']

# Mengatur ukuran figure dengan lebar 10 inch dan tinggi 5 inch
plt.figure(figsize=(10, 5))

# Membuat line chart dari setiap kolom dalam dataset dengan warna dan label yang sesuai
plt.plot(date, asimilasi, marker='o', color='blue', label='Asimilasi', linewidth=3)
plt.plot(date, observasi, marker='o', color='yellow', label='Observasi', linewidth=3)
plt.plot(date, non_asimilasi, marker='o', color='red', label='Non Asimilasi', linewidth=3)

# Mengatur nilai y dalam plot sampai 27
for num in [20, 21, 22, 23, 24, 25, 26, 27]:
  plt.axhline(y=num, color='gray', linestyle='-', linewidth=0.3)

# Mengatur format sumbu x untuk menampilkan jam
plt.gca().xaxis.set_major_formatter(dts.DateFormatter('%d %b %Y // %H:%M'))
plt.gca().xaxis.set_major_locator(dts.AutoDateLocator())
plt.xticks(rotation=50, ha='right')

# Memberi keterangan pada gambar
plt.title('Suhu Permukaan', fontsize=15)
plt.xlabel('Waktu (UTC)', fontsize=14)
plt.ylabel('Celcius', fontsize=14)
plt.legend(loc='upper center', bbox_to_anchor=(1.1, 0.3))
plt.grid(True, axis='y', linewidth='0.2', color='gray', zorder=0)

# Menampilkan line chart
plt.show()
```
